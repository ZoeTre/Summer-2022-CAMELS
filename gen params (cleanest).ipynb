```python
def give_parameters(LineNumber):
    
    sorted_parameters = []

    parameter_set = ''

    for num in nw_st.rsplit(', '):
    
        parameter_set += (num + ', ')
    
        if len(parameter_set.rsplit(', ')) == 6:
        
            sorted_parameters.append(parameter_set)
        
            parameter_set = ''
            
    return (sorted_parameters[LineNumber])
```
